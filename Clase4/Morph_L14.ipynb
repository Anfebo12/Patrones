```
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

# Taking the file
img = cv2.imread('/content/drive/My Drive/Docencia/Patrones/Clases/Montadas/C4/j.png',0)
cv2_imshow(img)
```

```
# Erosion
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)

cv2_imshow(erosion)
```

```
# Dilatacion
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)

cv2_imshow(dilation)
```

```
# Gradientes morfologicos
kernel = np.ones((3,3),np.uint8)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

cv2_imshow(gradient)
```

```
img = cv2.imread('/content/drive/My Drive/Docencia/Patrones/Clases/Montadas/C4/j2.png',0)

# Apertura
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

_ , axes = plt.subplots(1,2, figsize=(4,4))
axes[0].imshow(img, cmap='gray'); axes[0].set_xticks([]); axes[0].set_yticks([])
axes[1].imshow(opening, cmap='gray'); axes[1].set_xticks([]); axes[1].set_yticks([])
```

```
img = cv2.imread('/content/drive/My Drive/Docencia/Patrones/Clases/Montadas/C4/j3.png',0)

# Cierre
kernel = np.ones((5,5),np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

_ , axes = plt.subplots(1,2, figsize=(4,4))
axes[0].imshow(img, cmap='gray'); axes[0].set_xticks([]); axes[0].set_yticks([])
axes[1].imshow(closing, cmap='gray'); axes[1].set_xticks([]); axes[1].set_yticks([])
```
