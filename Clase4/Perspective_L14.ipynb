# **Are you sure of what you see? Playing with perspectives**

```
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt

# Take an image
image = cv2.imread("/content/drive/My Drive/Docencia/Patrones/Clases/Montadas/C3/sudoku1.jpg")
#cv2_imshow(image)
plt.figure(figsize = (15,10))
plt.imshow(image)
```

```
# Points center sudoku
points = np.float32([[370,170],[603,165],[362,387],[645,380]])
mapping = np.float32([[0,0],[300,0],[0,300],[300,300]])

#'''
# Validate selestec points
image = cv2.imread("/content/drive/My Drive/Docencia/Patrones/Clases/Montadas/C3/sudoku1.jpg")
cv2.circle(image, (370,170), 2, [0,0,255],3)
cv2.circle(image, (603,165), 2, [0,0,255],3)
cv2.circle(image, (362,387), 2, [0,0,255],3)
cv2.circle(image, (645,380), 2, [0,0,255],3)
plt.figure(figsize=(15,10))
plt.imshow(image)
#'''
```

```
Perspective = cv2.getPerspectiveTransform(points,mapping)

imagePerspective = cv2.warpPerspective(image, Perspective, (300,300))

plt.figure(figsize=(15,10)),plt.subplot(121),plt.imshow(image),plt.title('Input'), 
plt.figure(figsize=(10,5)),plt.subplot(122),plt.imshow(imagePerspective),plt.title('Output')
plt.show()
```
